```python
import glob
import nibabel as nb
import re
import pandas
```

```python
reg = re.compile('.*/registration/(?P<mask_type>.*)_masks/_mask_(?P<mask>.*)_subject_id_(?P<subject_id>.*)/_interp_(?P<interpolation>.*)/.*')
```

```python
# def get_volume(mask):
    
#     return (nb.load(mask).get_data() > 0).sum() * 1.96*1.96*2.75
    
```

```python
import subprocess
def get_volume(mask):
    output = subprocess.Popen(["fslstats", mask, "-V"], stdout=subprocess.PIPE).communicate()[0]
    return float(output.split()[1])
```

```python
def get_volume_weighted(mask):
    mask = nb.load(mask).get_data()
    return (mask[mask > 0]).sum() * 1.96*1.96*2.75
    
```

```python
df = []

for fn in glob.glob('/home/gdholla1/data/daphne/registration/*/_mask_*_subject_id_*/_interp_*/*.nii.gz'):
    df.append(reg.match(fn).groupdict())
    df[-1]['fn'] = fn
    df[-1]['volume'] = get_volume(fn)
    df[-1]['weighted_volume'] = get_volume_weighted(fn)
```

```python
df = pandas.DataFrame(df)
```

```python
df.pivot_table(index=['subject_id', 'mask'], columns=['mask_type', 'interpolation'],)['volume']
```
