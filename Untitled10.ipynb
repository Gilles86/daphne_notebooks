```python
def get_session_info(subject_id, run=3, shift=0):
    import pandas
    import numpy as np
    from nipype.interfaces.base import Bunch
    
    if (type(subject_id) == str) and  subject_id.startswith('S'):
        subject_id = int(subject_id[1:])

    df = pandas.read_pickle('/home/gdholla1/data/daphne/behavior/all_data3.pandas')
    
    df = df[(df.subject_id == subject_id)]

    df['onset'] += shift

    keys = ['standard.0', 'standard.1', 'oddball.0', 'oddball.1', 'shock']

    onsets = [df[df.condition == key].onset.tolist() for key in keys]

    info = Bunch(conditions=keys,
              onsets=onsets,
              durations=[[3]] * 5)
    
    return info
```
